```python
import csv
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

# Druglib Data

```python
Names = pd.read_csv("DruglibNames.txt")
link = 'http://www.druglib.com/ratingsreviews/'
add = "/"
Names['URL'] = Names.apply(lambda row: link + row.Names +  add, axis = 1) 
Total_comments = []
Total_ratings = []
Total_effectiveness = []
Total_sideeffects = []
Total_Benifts_review = []
Totatl_sideeffects_review = []
ratings = []
for i in range(len(Names)):
    url = Names.URL[i]
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html,"html.parser")
    table = soup.find('table', attrs={'width': '570'})
    for row in table.find_all('center'):
        for row2 in row:
            effectiveness = row.find_all('tr')[4].find_all('td')[2].text
            Total_effectiveness.append(effectiveness)
            sideeffects = row.find_all('tr')[5].find_all('td')[2].text
            Total_sideeffects.append(sideeffects)
            Benifts_review = row.find_all('tr')[12].find_all('td')[2].text
            Total_Benifts_review.append(Benifts_review)
            sideeffects_review = row.find_all('tr')[13].find_all('td')[2].text
            Totatl_sideeffects_review.append(sideeffects_review)
            comments = row.find_all('tr')[14].find_all('td')[2].text
            Total_comments.append(comments)
            rating = row.find_all('img', attrs={'src': '/img/red_star.gif'})
            c = 0
            for i in rating:
                c=c+1
            ratings.append(c)
drugLib_raw = pd.DataFrame(
        {'review': Total_comments,
         'rating': ratings,
         'effectiveness': Total_effectiveness,
         'sideEffects': Total_sideeffects,
         "benefitsReview": Total_Benifts_review,
         "sideEffectsReview":Totatl_sideeffects_review

        })
export_csv_drugLib_raw = drugLib_raw.to_csv (r'drugLib_raw.csv', index = None, header=True) 
drugLib_raw.head()
```

# Drugs.Com - Data

```python
Names = pd.read_csv("Drugs.ComNames.txt")
link = 'https://www.drugs.com/comments/'
add = "/"
Names['URL'] = Names.apply(lambda row: link + row.Names +  add, axis = 1) 
Total_comments = []
Total_ratings = []
for i in range(len(Names)):
    url = Names.URL[i]
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html,"html.parser")
    table = soup.find('div', attrs={'class': 'contentBox'})
    for row in table.find_all('div',attrs = {'class' : 'ddc-comment'})[1:]:
        comment = row.find_all('p',attrs = {'class' : 'ddc-comment-content'})[0].text
        Total_comments.append(comment)
        x = row.find_all('div',attrs = {'class' : 'comment-rating-score'})
        if(x == []):
            Total_comments.remove(comment)
        else:
            y = x[0].text
            Total_ratings.append(y)
drugsCom_raw = pd.DataFrame(
        {'review': Total_comments,
         'rating': Total_ratings
        })
export_csv_drugsCom_raw = drugsCom_raw.to_csv (r'drugsCom_raw.csv', index = None, header=True) 
drugsCom_raw.head()
```
